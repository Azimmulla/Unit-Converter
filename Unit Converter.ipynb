```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

def feet_to_meters(feet):
    return feet * 0.3048

def meters_to_feet(meters):
    return meters / 0.3048

def temperature_converter():
    print("Temperature Converter:")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")

    choice = input("Enter your choice (1 or 2): ")

    if choice not in ['1', '2']:
        print("Invalid choice. Please enter either 1 or 2.")
        return

    try:
        value = float(input("Enter the temperature value: "))
    except ValueError:
        print("Invalid input. Please enter a numeric value.")
        return

    if choice == '1':
        result = celsius_to_fahrenheit(value)
        print(f"{value} Celsius is equal to {result} Fahrenheit.")
    else:
        result = fahrenheit_to_celsius(value)
        print(f"{value} Fahrenheit is equal to {result} Celsius.")

def length_converter():
    print("Length Converter:")
    print("1. Feet to Meters")
    print("2. Meters to Feet")

    choice = input("Enter your choice (1 or 2): ")

    if choice not in ['1', '2']:
        print("Invalid choice. Please enter either 1 or 2.")
        return

    try:
        value = float(input("Enter the length value: "))
    except ValueError:
        print("Invalid input. Please enter a numeric value.")
        return

    if choice == '1':
        result = feet_to_meters(value)
        print(f"{value} Feet is equal to {result} Meters.")
    else:
        result = meters_to_feet(value)
        print(f"{value} Meters is equal to {result} Feet.")

# Main program
print("Welcome to the Unit Converter!")
print("Select an option:")
print("1. Temperature Converter")
print("2. Length Converter")

main_choice = input("Enter your choice (1 or 2): ")

if main_choice == '1':
    temperature_converter()
elif main_choice == '2':
    length_converter()
else:
    print("Invalid choice. Please enter either 1 or 2.")
```
